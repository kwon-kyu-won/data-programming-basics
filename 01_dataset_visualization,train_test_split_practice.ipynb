```python
import mglearn
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
#generate dataset
X,y = mglearn.datasets.make_forge()
#plot dataset
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.legend(["Class 0", "Class 1"], loc=4) # 범례 추가
plt.xlabel("First feature")              # x축 이름
plt.ylabel("Second feature")             # y축 이름
print("X.shape: {}".format(X.shape))     # 데이터 구조 출력
```

```python
#test_size=7: 테스트용 데이터를 정확히 7개만 뽑겠다는 뜻.
#random_state=2021: 무작위로 섞을 때 난수시드 고정. 이 숫자를 똑같이 쓰면 다음에 실행해도 항상 같은 데이터들로 나뉨.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=7, random_state=2021)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
print(y_test)
```

```python
#test_size=0.2: 정수가 아닌 소수를 넣으면 비율(20%)을 의미
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2021)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
print(y_test)
```

```python
#stratify =y: 타겟 데이터(y)의 클래스 비율(예: 0이 50%, 1이 50%)을 훈련 세트와 테스트 세트에서도 그대로 유지하도록 나눔.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=2021)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
print(y_test)
```
